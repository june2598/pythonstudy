```python
import pandas as pd
```

```python
dict1 = {'name':['홍길동', '홍길동', '김철수', '박영희', '김철수', '김철수', '홍길동'], 'korean': range(7)}
df1 = pd.DataFrame(dict1)
df1
```

```python
dict2 = {'name':['김철수', '홍길동', '심수봉'], 'english':range(3)}
df2 = pd.DataFrame(dict2)
df2
```

```python
pd.merge?
```

```python
# how의 default값 : inner
# on의 default값 : none => 같은게 있으면 자동으로 찾음 => 밑의 문장에서는 on='name'이 그래서 굳이 필요가 없음
pd.merge(df1, df2, on='name')
```

```python
print("\n# how='outer' 라고 명시하면 full outer join이다.")
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
# 충돌하는 column 값이 없어서 나타나진 않음
pd.merge(df1, df2, how='right', suffixes = ['_x','_y'])
```

```python
dict3 = {'leftkey':['홍길동', '홍길동', '김철수', '박영희', '김철수', '김철수', '홍길동'], 
         'korean': range(7)}
df3 = pd.DataFrame(dict3)
df3
```

```python
dict4 = {'rightkey':['김철수', '홍길동', '심수봉'], 'english':range(3)}
df4 = pd.DataFrame(dict4)
df4
```

```python
#도메인이 같고, 컬럼명이 다른 형태의 조인
pd.merge( df3, df4, left_on='leftkey', right_on='rightkey').drop('rightkey',axis=1)
```

```python
#도메인이 같고, 컬럼명이 다른 형태의 조인
pd.merge( df3, df4, left_on='leftkey', right_on='rightkey')
```

```python
dict1 = {'key1':['김철수', '김철수', '박영희'], 
         'key2':['one', 'two', 'one'], 
         'leftval':[1, 2, 3]}
left = pd.DataFrame(dict1)
left
```

```python
dict2 = {'key1':['김철수', '김철수', '박영희', '박영희'], 
         'key2':['one', 'one', 'one', 'two'], 
         'leftval':[4, 5, 6, 7]}
right = pd.DataFrame(dict2)
right
```

```python
# 여러 개의 컬럼을 병합하려면, 컬럼 이름이 들어간 리스트를 넘겨 주면 됩니다.
mylist = ['key1', 'key2'] # 조인할 컬럼 리스트

#  여러 개의 컬럼 병합하기
pd.merge(left, right, on=mylist, how='outer', suffixes=('_1','_2'))
```

```python
pd.merge(left, right, on='key1', suffixes=('_왼쪽', '_오른쪽'))
```

```python
newdf1 = df1.reset_index(drop=True)
newdf1
```

```python
newdf2 = df2.set_index('name')
newdf2
```

```python
# left_index=True:
#     newdf1의 인덱스를 병합 기준으로 사용합니다.
# right_index=True:
#     newdf2의 인덱스를 병합 기준으로 사용합니다.
# indicator=True:

#     병합 결과에 각 행이 어떤 데이터프레임에서 유래했는지를 나타내는 _merge라는 새로운 열을 추가합니다. 이 열은 다음과 같은 값을 가질 수 있습니다:
#         'left_only': newdf1에만 존재하는 인덱스
#         'right_only': newdf2에만 존재하는 인덱스
#         'both': 두 데이터프레임 모두에 존재하는 인덱스
```

```python
result = pd.merge(newdf1, newdf2, left_index=True, right_index=True, how='outer', indicator = True)
result
```
