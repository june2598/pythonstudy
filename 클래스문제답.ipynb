- 예외 처리

```python
class Book :
  def __init__(self,title,author,isbn) :
    self.title = title
    self.author = author
    self.isbn = isbn
    self.available = True
  def __str__(self) :
    return f'제목:{self.title},저자:{self.author},ISBN:{self.isbn},대여가능:{self.available}'

class Library :

  def __init__(self) :
    self.books = []

  # 도서 추가
  def add_book(self,book) :
    for ele in self.books :
      if ele.isbn == book.isbn :
        # print('isbn 중복!')
        raise ValueError('isbn 중복!')
        return

    self.books.append(book)
    print(f'{book.title}가 추가됨!')

  #도서 제거
  def remove_book(self,isbn) :
    for ele in self.books :
      if ele.isbn == isbn :
        self.books.remove(ele)
        print(f'{ele.title}가 제거됨!')
        return
    # print(f'ISBN = {isbn} 도서를 찾을수 없음!')
    raise ValueError(f'ISBN = {isbn} 도서를 찾을수 없음!')

  #도서 검색
  def find_book(self,isbn) :
    for ele in self.books :
      if ele.isbn == isbn :
        return ele
    # print(f'ISBN = {isbn} 도서를 찾을수 없음!')
    raise ValueError(f'ISBN = {isbn} 도서를 찾을수 없음!')

  #도서 목록
  def list_books(self) :
    for ele in self.books :
      print(ele)
```

```python
#도서 생성
book1 = Book('파이썬 프로그래밍','홍길동', '1')
book2 = Book('자바 프로그래밍', '김철수', '2')
```

```python
#도서관 생성
library = Library();
```

```python
#도서 추가
library.add_book(book1)
library.add_book(book2)
```

```python
#도서 목록
library.list_books()
```

```python
#도서 추가(ISBN중복)
book3 = Book('데이터베이스','홍길순','3')
try :
  library.add_book(book3)
except ValueError as e:
  print(e)
else :
  print(f'ISBN:{book3.isbn} 도서가 추가되었습니다.')
```

```python
#도서 검색
finded_book=None
try :
  print(library.find_book('4'))
except ValueError as e:
  print(e)
else :
  print(finded_book)
```
