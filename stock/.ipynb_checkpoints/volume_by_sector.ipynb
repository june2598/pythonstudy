특정 섹터의 코드를 리스트로 가져와서

섹터의 리스트를 순회하면서 

sector별 종목코드를 리스트로 가져옴

가져온 종목코드 리스트를 for문을 기반으로-> sector 내 종목의 date와 volume만을 가져옴

같은 sector 내 종목간의 volume을 더해, 각 날짜별 그 sector의 volume 총합을 구함

이를 통해 특정 sector의 기간별 volume의 변화를 파악-> 그래프로 시각화

```python
# fdr.StockListing('ETF/KR')

etfs = fdr.StockListing('ETF/KR')
# 파생열 추가 CategoryName
etfs['CategoryName'] = etfs['Category']
# Category코드를 코드명으로 변경

# 1. 국내 시장지수 2. 국내 업종/테마 3. 국내파생 4. 해외주식 5. 원자재 6. 채권 7. 기타
category_decode = ['국내 시장지수','국내 업종/테마' ,'국내파생' ,'해외주식','원자재','채권','기타']
etfs['CategoryName'] = etfs['CategoryName'].replace([i+1 for i in range(7)],category_decode)
columns = ['Symbol','Category','CategoryName','Name','Price','RiseFall','Change','ChangeRate','NAV','EarningRate','Volume','Amount','MarCap']
# case1) 열 재배치
etfs = etfs[columns]
etfs.head(2)
```

```python

# sector별 종목코드를 리스트로 가져옴
category_symbols = etfs.groupby('CategoryName')['Symbol'].agg(list)

# CategoryName을 key로, 종목코드의 list를 value로 갖는 딕셔너리
category_symbols_dict = category_symbols.to_dict()



# 오늘 날짜 구하기
today = datetime.datetime.today()
today_str = today.strftime('%Y-%m-%d')

# n개월 전 날짜 계산 함수
def calculate_start_date(months_ago, end_date):
    start_date = datetime.datetime.strptime(end_date, '%Y-%m-%d') - relativedelta(months=months_ago)
    return start_date.strftime('%Y-%m-%d')

end_date = today_str  # 오늘 날짜


# 섹터 이름별 n개월 Volume 추이를 보는 함수
# month_ago : 몇개월 전까지 볼지, (default=1개월)
# end_date : 검색 종료일, (default=오늘)

def volume_by_sector_name(category_name, months_ago=1, end_date=today_str):
    
    start_date = calculate_start_date(months_ago, end_date)
    
    # 새로운 데이터프레임 생성
    new_data = pd.Series(dtype='float64')  # 빈 시리즈로 초기화

    for code in category_symbols_dict.get(category_name, []):     #존재하지 않는 카테고리명일 경우 빈리스트 반환
        try:
            # 데이터 가져오기 (Volume만 선택)
            data = fdr.DataReader(code, start=start_date, end=end_date)[['Volume']]
            # Volume 합산
            new_data = new_data.add(data['Volume'], fill_value=0)  # fill_value를 사용하여 NaN 처리
            
        except Exception as e:
            print(f"{code} 데이터 가져오기 오류: {e}")
    
    return new_data
  
# 특정 섹터의 거래량 추이 예시
sector_volume = volume_by_sector_name('국내 시장지수')
type(sector_volume)
```

```python
# 선 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(sector_volume.index, sector_volume.values, marker='o')
plt.title('섹터별 거래량 추이 - 국내 시장지수')
plt.xlabel('날짜')
plt.ylabel('거래량')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 모든 섹터의 거래량 추이를 비교하는 함수
def volume_by_all_sectors(months_ago=1, end_date=today_str):
    start_date = calculate_start_date(months_ago, end_date)

    # 섹터이름을 키로 가져가야 되기때문에, 빈 리스트가 아닌 빈 딕셔너리로 시작
    sector_volumes = {}

    for category_name, symbols in category_symbols_dict.items():
        new_data = pd.Series(dtype='float64')  # 빈 시리즈로 초기화
        
        for code in symbols:
            try:
                # 데이터 가져오기 (Volume만 선택)
                data = fdr.DataReader(code, start=start_date, end=end_date)[['Volume']]
                # Volume 합산
                new_data = new_data.add(data['Volume'], fill_value=0)
            except Exception as e:
                print(f"{code} 데이터 가져오기 오류: {e}")
        
        sector_volumes[category_name] = new_data  # 각 섹터의 거래량 저장

    return sector_volumes

# 모든 섹터의 거래량 데이터 가져오기
all_sector_volumes = volume_by_all_sectors(months_ago=1)

# 그래프 그리기
plt.figure(figsize=(14, 8))

for sector_name, volume_data in all_sector_volumes.items():
    plt.plot(volume_data.index, volume_data.values, marker='o', label=sector_name)

plt.title('모든 섹터의 거래량 추이 비교')
plt.xlabel('날짜')
plt.ylabel('거래량')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# stateless 방식으로 각 섹터별 거래량을 각각의 axes에 그림

num_sectors = len(all_sector_volumes)
fig, axes = plt.subplots(nrows=num_sectors, ncols=1, figsize=(10, 5 * num_sectors), sharex=True)

for ax, (sector_name, volume_data) in zip(axes, all_sector_volumes.items()):
    ax.plot(volume_data.index, volume_data.values, marker='o', label=sector_name)
    ax.set_title(sector_name)
    ax.set_ylabel('거래량')
    ax.grid()
    ax.legend()

plt.xlabel('날짜')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
