```python

def get_number():
  try:
    num=int(input('숫자를 입력하세요:'))
    result = 10 / num
  except:
    print(f'예외발생')
  else:
    print('예외가 발생하지 않을경우 수행됨')
    print(f'결과:{result}')
  finally:
    print('예외 유무 상관없이 수행됨')
```

```python
get_number()
```

```python
def divide_numbers():
  try:
    num1 = int(input('첫번째 숫자 : '))
    num2 = int(input('두번째 숫자 : '))
    result = num1 / num2
  except ValueError as e1:
    print(f'예외발생 Value Error')
    print(type(e1))
    print(e1.args)
    print(e1)     # e1.args 튜플의 첫번쨰 요소, 즉 에러메시지
  except ZeroDivisionError as e2:
    print(f'예외발생 ZeroDivisionError')
    print(type(e2))
    print(e2.args)
    print(e2)
  else:
    print(f'결과 : {result}')
  finally:
    print('오류 유무 상관없이 출력')
```

```python
def divide_numbers():
  try:
    num1 = int(input('첫번째 숫자 : '))
    num2 = int(input('두번째 숫자 : '))
    result = num1 / num2
  except (ValueError, ZeroDivisionError) as e:
    print(f'예외발생 Value Error')
    print(type(e))
    print(e.args)
    print(e)     # e1.args 튜플의 첫번쨰 요소, 즉 에러메시지
  else:
    print(f'결과 : {result}')
  finally:
    print('오류 유무 상관없이 출력')
```

```python
divide_numbers()
```
