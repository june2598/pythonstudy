# 문제 1:매개변수의 타입 검사

- 문제 : 두개의 숫자를 받아서 그 합계를 반환하는 함수를 작성하되, 입력된 매개변수가 숫자인지 확인하고, 숫자가 아닐경우 ValueError를 발생 시키도록하세요.

```python
def sum_number(x, y) :
  if not isinstance(x, (int|float)) or not isinstance(y, (int|float)) :
    raise ValueError('숫자를 입력해주세요.')
  return x + y
```

```python
result = sum_number(1,3)
print(result)
```

# 문제 2: 기본값과 가변 매개변수 혼합

- 문제: 기본값을 가지는 매개변수와 가변 매개변수를 사용하여, 특정 문자열을 여러번 반복하는 함수를 작성하세요. 기본값은 1로 설정하고, 가변매개변수로 추가 문자열을 받을수 있어야 합니다.

```python
def repeat_string(string, n=1, *args) :
  result = string * n
  for arg in args :
    result += arg
  return result
```

```python
# 테스트 예시
print(repeat_string('kh',3))
print(repeat_string('kh',3,'인재교육원','kdt'))
print(repeat_string('kh'))
```

#문제 3: 키워드 전용 매개변수와 기본값

- 문제 : 두 개의 숫자를 받아서 그 곱을 계산 하는 함수를 작성하되, 두번쨰 숫자는 키워드 전용 매개변수로 설정하고, 기본값은 1로 설정하세요.

# 문제 4: 매개변수의 언패킹

- 문제: 딕셔너리를 받아서 키와 값을 출력하는 함수를 작성하세요. 이 함수는 키워드 매개변수를 사용하여 딕셔너리의 내용을 언패킹해야 합니다.

```python
def unpacked_dic(**data) :
  for key, value in data.items():
    print(f'{key}: {value}')
```

```python
unpacked_dic(a=1, b=2, c=3)
```

```python
person = {'name':'홍길동', 'age':30}
unpacked_dic(**person)
```

```python
unpacked_dic(**{'name':'홍길동','age':30})
```

# 문제5: 위치 매개변수와 키워드 전용 매개변수 혼합 사용하기

- 문제: 두 개의 숫자와 연산자(문자열)를 받아서 계산결과를 반환하는 함수를 작성하되, 연산자는 키워드 전용 매개변수로 작성


```python
def carculate(num1, num2, *, operator) :
  match operator:
    case '+':
      return num1+num2
    case '-':
      return num1-num2
    case '*':
      return num1*num2
    case '/':
      if num2 == 0
        raise ValueError('0으로 나눌수 없습니다.')
      return num1/num2
    case _:
      raise ValueError('올바른 연산자를 입력해주세요.')
```

```python
carculate(3, 2, operator='^')
```

```python
def carculate2(num1, num2, *, operator) :
  if operator== '+' :
    return num1+num2
  elif operator =='-':
    return num1-num2
  elif operator == '*':
    return num1*num2
  elif operator == '/':
    if num2 == 0
      raise ValueError('0으로 나눌수 없습니다.')
    return num1/num2
  else :
    raise ValueError('올바른 연산자를 입력해주세요.')
```

```python
carculate2(3, 2, operator='*')
```

```python
carculate2(3,2,operator='^')
```
