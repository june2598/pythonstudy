```python
from abc import abstractmethod
class Person() :
  def __init__(self) :
    raise Exception('추상클래스')

  @abstractmethod
  def sleep(self):
    pass

  def smile(self) :
    print('웃다')

  def eat(self) :
    print('먹다')
```

```python
# Person 타입의 객체 생성 불가
p = Person('홍길동',20)
```

```python
class Student(Person) :
  def __init__(self,name,age,major) :
    # super().__init__(name,age)    #super() : 부모의 멤버에 접근할때
    self.name = name
    self.age = age
    self.major = major

  def sleep(self) :
    print(f'{self.name} 가 잠을 잡니다.')

  def eat(self) :
    print(f'{self.name} 가 먹다')
```

# getter, setter
- 파이썬에서는 getter,setter를 메소드호출 방식이 아니라 속성 접근하듯이 사용가능

```python
class Person :
  def __init__(self,name,age) :
    self._name = name
    self._age = age

  @property
  def name(self) :
    return self._name + '님'

  @name.setter
  def name(self,name) :
    self._name = name
```

```python
p1 = Person('홍길동',20)
print(p1.name)
p1.name = '홍길순'
print(p1.name)
```

```python
p1 = Person('홍길동',20)
p1.get_name()
p1.set_name('홍길순')
p1.get_name()
```
