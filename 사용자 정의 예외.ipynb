# 사용자 정의 예외 클래스
- Exception 클래스를 상속받아서 만들 수 있습니다.

- 예제1) 나이 검증 예외

나이가 0보다 작거나 120보다 큰 경우 예외발생

```python
class MyCustomError(Exception):
  pass
```

```python
class AgeError(Exception):
  pass

def set_age(age):
  if age < 0 or age > 120:
    raise AgeError("나이는 0 이상 120 이하의 값이어야 합니다.")
  return age

try:
  set_age(-1)
except AgeError as e:
  print(e)
```

- 예제2: 비밀번호 길이 예외

비밀번호가 8자 미만일 경우 예외를 발생시키는 예제

```python
class PasswordTooShortError(Exception):
  pass

def set_password(password):
  if len(password) < 8:
    raise PasswordTooShortError("비밀번호는 최소 8자 이상이어야 합니다.")
  return password

try:
  set_password("12345")
except PasswordTooShortError as e:
  print(e)
```
