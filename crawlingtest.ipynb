```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://finance.naver.com/sise/sise_group.naver?type=upjong'
res = requests.get(url)
print(res.status_code)
```

```python
soup = BeautifulSoup(res.content,'html.parser')
```

```python
soup
```

```python
soup.select('#contentarea_left > table > tr > td.tc span')
```

```python

graph_data = []
graph_span = soup.select('#contentarea_left > table > tr > td.tc span')
for span in graph_span:
  graph_data.append(span.text.strip())

print(graph_data)
```
