```python
# 섹터간 Volume(거래량)변동성을 비교

# 오늘하루 섹터별 거래량, 표준편차
category_stats = df_etfs.groupby('CategoryName')['Volume'].agg({'mean', 'std'}).reset_index()
category_stats.plot()

plt.figure(figsize=(10, 6))
plt.bar(category_stats['CategoryName'], category_stats['mean'], yerr=category_stats['std'], capsize=5)
plt.title('섹터별 거래량 및 표준편차')
plt.xlabel('섹터')
plt.ylabel('거래량')
plt.show()


```

```python
category_stats['Normalized'] = (category_stats['mean'] - category_stats['mean'].min()) / (category_stats['mean'].max() - category_stats['mean'].min())

plt.figure(figsize=(10, 6))
plt.bar(category_stats['CategoryName'], category_stats['Normalized'])
plt.title('정규화된 섹터별 거래량')
plt.xlabel('섹터')
plt.ylabel('정규화된 거래량')
plt.show()
```

```python
# 특정 ETF 데이터 가져오기
etf_code = 'SPY'

# 특정 ETF 데이터 6개월간 추이
etf_data = fdr.DataReader(etf_code, start=start_date, end=end_date)
```

```python
etf_data.head()
```

1. Open: 시작가. 거래 시작 시의 가격입니다. 해당 거래일의 첫 거래 가격을 나타냅니다.

2. High: 고가. 거래일 동안의 최고 가격입니다. 해당 거래일에 기록된 가장 높은 가격을 나타냅니다.

3. Low: 저가. 거래일 동안의 최저 가격입니다. 해당 거래일에 기록된 가장 낮은 가격을 나타냅니다.

4. Close: 종가. 거래 종료 시의 가격입니다. 해당 거래일의 마지막 거래 가격을 나타냅니다.

5. Volume: 해당 거래일 동안 거래된 주식의 수입니다. 거래량을 나타냅니다.

6. Adj Close: 조정 종가. 배당금, 주식 분할 등으로 인한 가격 조정을 반영한 가격으로, 투자자가 실제로 얻는 수익률을 계산하는 데 유용합니다.

```python
# 일별 수익률 계산
# pct_change()는 각 항목이 이전 항목에 비해 얼마나 변화했는지를 백분율로 계산하는 메소드
etf_data['Return'] = etf_data['Close'].pct_change()
```

```python
# 주별 수익률 계산
etf_weekly = etf_data.resample('W').agg({'Close': 'last'})
etf_weekly['Weekly Return'] = etf_weekly['Close'].pct_change()

# 월별 수익률 계산
etf_monthly = etf_data.resample('M').agg({'Close': 'last'})
etf_monthly['Monthly Return'] = etf_monthly['Close'].pct_change()

# 거래량 변동
etf_data['Volume'] = etf_data['Volume']
```

```python
# 시각화
plt.figure(figsize=(14, 12))

# 일별 수익률 시각화
plt.subplot(3, 1, 1)
plt.plot(etf_data.index, etf_data['Return'], label='Daily Return', color='blue')
plt.title('Daily Return of ETF')
plt.xlabel('Date')
plt.ylabel('Return')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()


# 주별 수익률 시각화
plt.subplot(3, 1, 2)
plt.plot(etf_weekly.index, etf_weekly['Weekly Return'], label='Weekly Return', color='orange')
plt.title('Weekly Return of ETF')
plt.xlabel('Date')
plt.ylabel('Return')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()

# 거래량 시각화
plt.subplot(3, 1, 3)
plt.bar(etf_data.index, etf_data['Volume'], color='green', alpha=0.6)
plt.title('Trading Volume of ETF')
plt.xlabel('Date')
plt.ylabel('Volume')

plt.tight_layout()
plt.show()
```
