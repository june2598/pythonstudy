```python
file_name = 'D:/kdt/project/crawling/sector_info_df_2024-12-18.csv'
```

```python
df = pd.read_csv(file_name)
```

```python
df.head(5)
```

```python
df['전일비'] = df['전일비'].str.replace('%','').astype(float)
```

```python
sorted_df = df.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(sorted_df['종목명'],sorted_df['전일비'], color='skyblue')
plt.title('상위 10 업종 전일비 추이', fontsize=16)
plt.xlabel('업종명', fontsize=14)
plt.ylabel('전일비(%)', fontsize=14)
plt.xticks(rotation=45)
plt.ylim(0, sorted_df['전일비'].max() + 5) # y축범위 설정
plt.grid(axis='y')
plt.show()
```
