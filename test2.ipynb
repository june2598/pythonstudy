```python
def describe_shape(shape):
  match shape:
    case 'triangle':
      print('삼각형입니다.')
    case 'circle':
      print('원입니다.')
    case 'rectangle':
      print('사각형입니다.')
    case _ :
      return "Unknown shape"
describe_shape('circle')
```

```python
def describe_shape(shape, *args):
  match (shape, *args):
    case ('triangle',base ,height):
      print('삼각형입니다.')
    case ('circle',radius):
      print('원입니다.')
    case ('rectangle',width,height):
      print('사각형입니다.')
    case _ :
      return "Unknown shape"
describe_shape('circle',5)
describe_shape('triangle',1,3)
describe_shape('rectangle',10,20)
```

## 가변 위치 매개 변수
- 임의의 수의 위치 인자를 튜플 형태로 받음.
- 함수 내에서 args로 접근할 수 있으며 *args는 해당 함수의 매개변수 목록에서 마지막 위치 매개변수 이후에 와야함.

```python
def sum_numbers(*args):
  result = 0
  for number in args:
    result += number
  return result


print(sum_numbers(1, 2, 3, 4))
```

## 가변 키워드 매개변수
- 임의의 수의 키워드 인자를 딕셔너리 형태로 받음.
- 함수 내내에서 args로 접근할수 있으며, **args는 매개변수 목록의 맨 마지막에 위치해야 합니다.

```python
def f(**args) :
  print(type(args), args)
  for k,v in args.items() :
    print(f'키={k} 값={v}')
```

```python
f(v1= '20', v2=30, v3=40)
```

```python
f(**d)
```

```python
def f( lab, numbers) :
  result = []
  for x in numbers :
    result.append(lab(x))
  return result
```

```python
numbers = [1,2,3]
result = f(lambda x: x**2, numbers)
print(result)
```

## 이중 언패킹
- 딕셔너리를 언팩할 떄 사용. 이를 통해 딕셔너리의 키-값 쌍을 개별적으로 분리하거나, 함수의 키워드 인자로 전달할 수 있음

```python
data = {'name' : 'Alice', 'age' : 30}
print(data['name'])
```

```python
def greet(name, age):
  print(f"Hello, my name is {name} and I am {age} years old.")
```

```python
greet(**data)
```
