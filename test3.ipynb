```python
# 34. 'Quantity'가 3 이상인 주문의 평균 가격을 구하세요.

display(df[df['Quantity'] >= 3]['Price'].mean())
display(df.loc[df['Quantity'] >= 3,'Price'].mean())
display(df.iloc[df[df['Quantity'] >= 3].index, df.columns.get_loc('Price')].mean())
display(df.query("Quantity >= 3")['Price'].mean())
```

```python
# 35. 'Product'별 판매량을 시계열로 분석하세요.
# case1)
# df.pivot_table(index='Date',columns='Product',values='Quantity',aggfunc='sum')

# case2)
# df.groupby(['Date','Product']).agg({'Quantity':'sum'}).reset_index().pivot(index='Date',columns='Product',values='Quantity')

# step1)
print(df.groupby(['Date','Product']).agg({'Quantity':'sum'}))
# step2)
print(df.groupby(['Date','Product']).agg({'Quantity':'sum'}).reset_index())
# step3)
print(df.groupby(['Date','Product']).agg({'Quantity':'sum'}).reset_index().pivot(index='Date',columns='Product',values='Quantity'))
```

```python
# 39. 'Date'별 가장 많이 팔린 제품을 구하세요.
# df.groupby(['Date','Product']).agg({'Quantity':'max'}).reset_index().drop('Quantity',axis=1)

df.groupby('Date').apply(lambda x:x.loc[x['Quantity'].idxmax()])['Product']
```

```python
# 40. 'CustomerID'별 구매 패턴을 분석하세요 (예: 구매 빈도, 평균 구매 수량).

df.groupby('CustomerID').agg({'OrderID':'count', 'Quantity':'mean'}) \
.rename(columns={'OrderID':'구매빈도','Quantity':'평균구매수량'}) \
.rename_axis('고객ID')
```
