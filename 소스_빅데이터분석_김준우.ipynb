```python
import pandas as pd
```

```python
# 1. 데이터셋을 불러와서 처음 5개의 행을 출력하시오

df = pd.read_csv('data.csv')
df.head()
```

```python
# 2. '카테고리' 별로 몇 개의 상품이 있는지 각각 세어 출력하시오

df['카테고리'].value_counts()
```

```python
# 3. 모든 상품의 평균 가격을 계산하시오
all_product_avg_price = df['가격'].mean()
print(f'모든 상품의 평균 가격 : {all_product_avg_price}원')
```

```python
# 4. '판매수량'이 가장 높은 상위 10개 상품을 출력하시오

df.sort_values('판매수량', ascending=False).head(10)
```

```python
# 5. 각 '카테고리'별 평균 판매수량을 계산하시오

df.groupby('카테고리').agg({'판매수량':'mean'})
```

```python
# 6. '판매날짜'를 기준으로 데이터를 정렬하시오(오래된 순)
df.sort_values('판매날짜',ascending=True)
```

```python
# 7. 각 '카테고리'별로 최고 가격을 가진 상품의 이름을 출력하시오.

max_price_idx = df.groupby(['카테고리'])['가격'].idxmax()

df.loc[max_price_idx,['카테고리','상품명','가격']]
```

```python
# 8. 모든 상품의 총 판매액(가격 * 판매수량)을 계산하시오

# 9. 각 상품별로 '판매액'을 새로운 파생열로 추가하시오
df['판매액'] = df['가격'] * df['판매수량']
display(df.head(10))

print(f'총 판매액 : {df['판매액'].sum()}')
```

```python
# 10. 각 '카테고리'별 총 판매액을 계산하시오

df.groupby('카테고리').agg({'판매액':'sum'}).rename(columns={'판매액':'총판매액'})
```

```python
# 11. 각 '카테고리'별 평균 판마액을 계산하고, 평균 판매액이 가장 높은 카테고리를 출력하시오.

display(df.groupby('카테고리').agg({'판매액':'mean'}).sort_values('판매액', ascending=False).head(1))
top_avg_price = df.groupby('카테고리').agg({'판매액':'mean'}).sort_values('판매액', ascending=False).head(1).index[0]
print(f'평균 판매액이 가장 높은 카테고리 : {top_avg_price}')
```

```python
# 12. 판매 데이터 중에서 '판매수량' 상위 25%에 해당하는 상품들만 선택하여, 그 상품들의 평균 '가격을 계산하시오

df[df['판매수량'] >= df['판매수량'].quantile(.75)].agg({'가격':'mean'})
```

```python
# 13. '판매날짜'를 datetime 타입으로 변환하고, 연도별 총 판매수량을 계산하시오.
df['판매날짜'] = pd.to_datetime(df['판매날짜'])
df.resample('YE', on='판매날짜').agg({'판매수량':'sum'})
```

```python
# 14. 각 카테고리 내에서 '가격' 상위 10%에 해당하는 상품들의 목록을 출력하시오.

top_10per_price = df.groupby('카테고리')['가격'].quantile(0.9).reset_index()
top_10per_price.columns = ['카테고리','상위10%가격']
display(top_10per_price)

new_df = df.merge(top_10per_price, on='카테고리', how='left')
# new_df

# 상위 10%에 해당하는 상품들의 목록
display(df[new_df['가격'] >= new_df['상위10%가격']].sort_values('카테고리'))


# 모범답안 
# top_10_percent_prices = df.groupby('카테고리').apply(lambda x: x[x['가격'] > x['가격'].quantile(0.9)])
# # print(top_10_percent_prices)
# display(top_10_percent_prices)
```

```python
# 15. '판매날짜'를 기준으로 각 달의 총 판매액을 계산하고, 판매액이 가장 높은 달을 출력하시오

display(df.resample('ME', on='판매날짜').agg({'판매액':'sum'}).sort_values('판매액', ascending=False).head(1))
top_sales_amount = df.resample('ME', on='판매날짜').agg({'판매액':'sum'}).sort_values('판매액', ascending=False).head(1).index[0].strftime('%m')


print(f'판매액이 가장 높은 달 : {top_sales_amount}월')


# 모범답안
# df['월'] = df['판매날짜'].dt.month
# monthly_sales = df.groupby('월')['판매액'].sum()
# print(monthly_sales)
# print(monthly_sales.idxmax(), '월')
```
