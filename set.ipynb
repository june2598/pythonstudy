## add() 단일 요소추가
- update() : 여러 요소 추가, iterable(리스트, 튜플,set, dictionary,문자열)
- remove() : 요소제거

```python
print(fruits)
fruits.add('mango')
```

- 이미 있는 값을 넣으면 중복으로 추가가안됨

```python
print(fruits)
fruits.add('apple')
print(fruits)
```

```python
fruits.update(['kiwi','mango'])
fruits
```

## 집합연산
- union(other_set) : 합집합, |
- intersection(other_set) : 교집합, &
- difference(other_set) : 차집합, -

```python
set1 = {1,2,3}
set2 = {3,4,5}
print(set1.union(set2))
print(set1 | set2)
```

```python
set1 = {1,2,3}
set2 = {3,4,5}
print(set1.intersection(set2))
print(set1 & set2)
```

```python
set1 = {1,2,3}
set2 = {3,4,5}
print(set1.difference(set2))
print(set1 - set2)
```
