- append() : 요소 추가
- extend() : 리스트를 항목별로 맨뒤에 추가
- insert() : 지정한 인덱스 위치에 삽입
- count() : 리스트의 갯수

```python
numbers = list(range(1,5+1))
numbers.append(6)
numbers
```

```python
numbers.append([7,8])
numbers
```

```python
numbers2 = list(range(10,50+1,10))
numbers2
```

```python
numbers.extend(numbers2)
numbers
```

```python
numbers = list(range(1,5+1))
numbers2 = list(range(10,50+1,10))
numbers3 = [*numbers, *numbers2]
numbers3
```

```python
numbers3.insert(numbers3.index(10),99)
numbers3
```

```python
numbers = [1,2,2,3]
numbers.count(2)
```

요소 삭제
- pop(index) : 맨뒤의 항목 반환 및 삭제, index를 지정하면 지정한 위치의 항목을 반환 및 제거
- remove(value) : 첫번째로 일치하는 값을 검색해서 제거, 삭제한 요소는 반환하지 않음

- del listData(index) : 리스트에서 해당 인덱스를 이용해 항목을 삭제

```python
numbers = list(range(1,10+1))
numbers
```

```python
return_value = numbers.pop()
print(return_value,numbers)
```

```python
return_value = numbers.pop(numbers.index(5))
print(return_value,numbers)
```

```python
numbers = [1,2,3,4,5,3]
return_value = numbers.remove(3)
print(return_value,numbers)
```

```python
numbers = [1,2,3,4,5]
del numbers[2]
print(numbers)
numbers.clear()
print(numbers)
```

정렬하기

- sort() : 정렬 (reverse=true 매개변수 지정하면 내림차순), 원본데이터 바뀜

- reverse() : 리스트의 항목들을 역순으로 나열(내림차순 정렬이 아님), 원본데이터 바뀜

- [::-1] : 역순으로 나열, 원본데이터를 바꾸지않음

```python
numbers = [3,2,1,5,4]
numbers.sort()
numbers
```

```python
numbers = [3,2,1,5,4]
numbers.sort(reverse=True)
numbers
```

```python
numbers = [3,2,1,5,4]
numbers.reverse()
numbers
```

```python
numbers = [3,2,1,5,4]
after_reverse = numbers[::-1]
print(after_reverse,numbers)
```

리스트 복제
- copy() : 리스트를 복사하여 새로운 리스트 반환

- =: 주소를 복사해 같은 객체를 참조

```python
numbers = [1,2,3,4,5]
new_numbers = numbers.copy()
print(new_numbers)
print(numbers is new_numbers)
print(id(numbers),id(new_numbers))
```

```python
numbers = [1,2,3,4,5]
new_numbers = numbers
print(new_numbers)
print(numbers is new_numbers)
print(id(numbers),id(new_numbers))
```
