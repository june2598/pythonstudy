```python
class Person:
  #정적 필드 : 바깥에 선언된. (객체를 생성하지 않아도)
  national = '한국'

  # 생성자 메소드
  def __init__(self, name, age):

  # 인스턴스 필드
    self._name = name
    self._age = age
    # national = '미국'   => national은 지역변수
    Person.national = '미국'    # 정적필드는 클래스명.필드명으로 접근

  # 인스턴스 메소드
  def smile(self) :
    print('웃다')
  def eat(self) :
    print('먹다')

  # 정적 메소드
  @staticmethod
  def staticMethod():
    print('정적메소드')

  # getter 메소드
  @property
  def name(self):
    return self._name

  @property
  def age(self):
    return self._age

  # setter 메소드
  @name.setter
  def name(self,name):
    self._name = name

  @age.setter
  def age(self,age):
    self._age = age

  def __str__(self):
    print(f'{self.name}-{self.age}')
```

```python
p1 = Person('홍길동',10)
p2 = Person('홍길순',20)
```

- 접근 방식: name은 @property 데코레이터로 정의된 getter 메서드를 통해 접근합니다. 이 메서드는 _name의 값을 반환합니다. 따라서 person.name을 호출하면 getter 메서드가 실행되어 _name의 값을 반환합니다.
- 의미: name은 비공식적인 접근 방식 대신 공식적인 인터페이스를 제공하며, 속성을 보호하고 추가적인 검증 로직을 추가할 수 있는 기회를 제공합니다. 예를 들어, setter 메서드를 통해 값을 설정할 때 유효성 검사를 수행할 수 있습니다.

```python
p1.name
```

- 접근 방식: _name은 인스턴스 변수로, 클래스 내부에서 직접 접근할 수 있습니다. 하지만, 일반적인 관례로 언더스코어(_)로 시작하는 변수는 "비공식"적으로 private 속성으로 간주됩니다. 즉, 클래스 외부에서 직접 접근하는 것을 피하는 것이 좋습니다.

- 의미: _name은 클래스 내부에서만 사용되는 데이터로, 외부에서 직접 수정하거나 접근하는 것을 피하고 싶을 때 사용합니다.

```python
class Parent:
  pass

class Child(Parent):
  pass
```

- issubclass() : 앞에 자식, 뒤에 부모. 결과값은 boolean 값으로 나옴

```python
print(issubclass(Child, Parent))
print(issubclass(Parent, Child))
```

- mro() => 메소드 해석 순서 (Method Resolution Order)

```python
class A:
  pass

class B(A):
  pass

class C(A):
  pass

class D(B,C):
  pass

print(D.mro())
```

```python
import inspect
```

```python
print(inspect.getmro(D))
```
